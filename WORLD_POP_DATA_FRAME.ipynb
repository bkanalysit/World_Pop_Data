```python
import numpy as np
import pandas as pd
#import matplotlid
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.rename(columns={'Country (or dependency)': 'Country'}, inplace=True)
df.rename(columns={'Population (2020)': 'Population'}, inplace=True)
df.rename(columns={'Yearly Change': 'Yearly_Change'}, inplace=True)
df.rename(columns={'Density (P/Km²)': 'Density_per_Km2'}, inplace=True)
df.rename(columns={'Land Area (Km²)': 'Land_Area_km'}, inplace=True)
df.rename(columns={'Fert. Rate': 'Fert_Rate'}, inplace=True)
df.rename(columns={'Med. Age': 'Med_Age'}, inplace=True)
df.rename(columns={'Urban Pop %': 'Urban_Pop_Pct'}, inplace=True)
df.rename(columns={'World Share': 'World_Share'}, inplace=True)
```

```python
df.replace("N.A.", np.nan, inplace=True)
```

```python
df = df.dropna(axis = "index", how = "any", subset = ["Population", "Density_per_Km2", "Land_Area_km", "Fert_Rate", "Med_Age", "Urban_Pop_Pct", "World_Share"])
```

```python
#df[["Yearly_Change", "New"]] = df["Yearly_Change"].str.split(" ", expand=True)
df[["Urban_Pop_Pct", "New_2"]] = df["Urban_Pop_Pct"].str.split(" ", expand=True)
df[["World_Share", "New_3"]] = df["World_Share"].str.split(" ", expand=True)
```

```python
df.drop(columns=["New", "New_2", "New_3"], inplace=True)
```

```python
df["Fert_Rate"] = df["Fert_Rate"].astype(float)
df["Med_Age"] = df["Med_Age"].astype(float)
df["Yearly_Change"] = df["Yearly_Change"].astype(float)
df["Urban_Pop_Pct"] = df["Urban_Pop_Pct"].astype(int)
df["World_Share"] = df["World_Share"].astype(float)
```
